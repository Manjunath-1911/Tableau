```python
corr = df.drop(columns=['Date']).corr()
print(corr)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Walmart Dataset")
plt.show()
```

```python
corr = df.drop(columns=['Date']).corr()
print(corr)
plt.figure(figsize=(8,6))
sns.scatterplot(corr)
plt.title("Correlation Heatmap of Walmart Dataset")
plt.show()
```
